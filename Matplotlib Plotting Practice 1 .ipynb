Now add in the vertical dashed average line with the corresponding annotation.

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv(r"D:\DEPI Python Jupyter\Matplotlib&Seaborn\Matplotlib & Seaborn_\Matplotlib\Practical exercise\MBA_Stats.csv")

fig, ax0 = plt.subplots()

n_bins = int(1 + 3.322 * len(df['Avg_GMAT']))
ax0.hist(df['Avg_GMAT'], bins=n_bins, edgecolor='black', color='skyblue')

ax0.set_xlabel('Average GMAT Score')
ax0.set_title('Histogram of Average GMAT Scores')

mean_gmat = df['Avg_GMAT'].mean()
ax0.axvline(mean_gmat, color='red', linestyle='--', linewidth=2)
ax0.text(mean_gmat + 5, ax0.get_ylim()[1] * 0.9, f'Mean: {mean_gmat:.2f}', color='red')

plt.show()
```

Now create the second histogram for work experience

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv(r"D:\DEPI Python Jupyter\Matplotlib&Seaborn\Matplotlib & Seaborn_\Matplotlib\Practical exercise\MBA_Stats.csv")

fig, ax1 = plt.subplots()

n_bins_job = int(1 + 3.322 * len(df['Avg_Job_Months']))
ax1.hist(df['Avg_Job_Months'], bins=n_bins_job, edgecolor='black', color='lightgreen')
ax1.set_xlabel('Average Job Experience (Months)')
ax1.set_title('Histogram of Average Job Experience')

mean_job = df['Avg_Job_Months'].mean()
ax1.axvline(mean_job, color='red', linestyle='--', linewidth=2)
ax1.text(mean_job + 5, ax1.get_ylim()[1] * 0.9, f'Mean: {mean_job:.2f}', color='red')

plt.show()
```

Create a 1X3 figure with a shared y axis and three axes. Set figsize = (10,10). Add the figure title to the figure.

```python
#Create axes + figure
#Create the plot on ax0, you should set kind = "barh". Add the appropriate titles and axes labels.
#Update axes features.  Make the x axis go from 600 to 690.
#Create the plot on ax1
#Update axes features.  Make the x axis go from  25 to 32.
#Create the plot on ax2
#Update axes features.  Make the x axis go from  45 to 75.

import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv(r"D:\DEPI Python Jupyter\Matplotlib&Seaborn\Matplotlib & Seaborn_\Matplotlib\Practical exercise\MBA_Stats.csv")

fig, (ax0, ax1, ax2) = plt.subplots(1, 3, figsize=(10, 10), sharey=True)

df.plot(kind='barh', y='Avg_GMAT', x='Industry', ax=ax0, legend=False, color='skyblue')
ax0.set_xlabel('Average GMAT Score')
ax0.set_ylabel('Industry')
ax0.set_title('Average GMAT by Industry')
ax0.set_xlim(600, 690)

df.plot(kind='barh', y='Avg_Age', x='Industry', ax=ax1, legend=False, color='lightgreen')
ax1.set_xlabel('Average Age')
ax1.set_title('Average Age by Industry')
ax1.set_xlim(25, 32)

df.plot(kind='barh', y='Avg_Job_Months', x='Industry', ax=ax2, legend=False, color='orange')
ax2.set_xlabel('Average Job Experience (Months)')
ax2.set_title('Average Job Experience by Industry')
ax2.set_xlim(45, 75)

fig.suptitle('Industry Statistics: GMAT, Age, and Job Experience', fontsize=16)

plt.tight_layout()
plt.show()
```
